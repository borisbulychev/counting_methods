# 1) x^3 - 2x - 5 = 0
## epsilon = 0.001
## start_point = 2.0



```python
from math import log10 as lg

def f(x):
  return x**3 - 2*x - 5

def f_pr(x):
  return 3*x**2 - 2

def new_meth(start_point, epsilon):
  x = start_point
  iters = 0
  x1 = x - f(x)/f_pr(x)
  while abs(x1-x)>epsilon:
    x = x1
    x1 = x - f(x)/f_pr(x)
    iters += 1
  return round(x1, int(abs(lg(epsilon)))+1), iters
start_point = 2.0
epsilon = 0.001
x_star, num_iters = new_meth(start_point, epsilon)
print (f": {x_star} c  {epsilon}.   - {num_iters}")
```

```python
import matplotlib.pyplot as plt
from decimal import Decimal as dec
from math import log10 as lg

def f(x):
  return x**3 - 2*x + 2

def f_pr(x):
  return 3*x**2 - 2

def tangent_line(x0, x_range):
    return [f_pr(x0) * (x - x0) + f(x0) for x in x_range]

def new_meth(start_point, epsilon, c):
  x = dec(start_point)
  iters = 0
  x1 = dec(x - c*dec((f(x)/f_pr(x))))
  plt.plot(x_values, tangent_line(x, x_values))
  plt.plot(x_values, tangent_line(x1, x_values))
  while abs(x1-x)>epsilon and iters < 30:
    x = x1
    x1 = x - c*(f(x)/f_pr(x))
    plt.plot(x_values, tangent_line(x1, x_values))
    iters += 1
  return round(float(x1), int(abs(lg(epsilon)))+1), iters

start_point = 0.0
epsilon = 0.001
x_values = [dec(i / 100) for i in range(-500, 300)]

for c in range(-1000,-990):
  plt.figure(figsize=(8, 6))
  plt.plot(x_values, [f(x) for x in x_values], color='blue')
  x_star, num_iters = new_meth(start_point, epsilon, dec(c))
  plt.axhline(0, color='black',linewidth=0.5)
  plt.axvline(0, color='black',linewidth=0.5)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.grid(True)
  plt.show()
  print (f": {x_star} c  {epsilon}.   - {num_iters}", c)
```
